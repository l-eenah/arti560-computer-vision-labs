### Import Libraries


```python
import tensorflow as tf
from tensorflow import keras 
from keras import layers
import keras_hub #provides pretrained models like: ViT, Bert # pyright: ignore[reportMissingImports]
import tensorflow_datasets as tfds # pyright: ignore[reportMissingImports]
import numpy as np
import os

# Hyperparameters
image_size = 224 #image size for the preset model
batch_size = 32
num_classes = 5
epochs = 5
learning_rate = 1e-4
```

### Load ViTImageClassifierPreprocessor (vit_base_patch16_224_imagenet preset)


```python
preprocessor = keras_hub.models.ViTImageClassifierPreprocessor.from_preset(
    "vit_base_patch16_224_imagenet"
)
```

### Load ViTImageClassifier (vit_base_patch16_224_imagenet preset)


```python
classifier = keras_hub.models.ViTImageClassifier.from_preset(
    "vit_base_patch16_224_imagenet",
    preprocessor=preprocessor
)
```

### Load the images

```python
## chosen_classes = ['kite','African_crocodile','jellyfish','ladybug','llama']

input_path= r'/content/test_images'
```

### Predict classes


```python
mapping={
    'kite':21,
    'African_crocodile':49,
    'jellyfish':107,
    'ladybug':301,
    'llama':355,
}
```

```python
from PIL import Image

for fname in os.listdir(input_path):
    image = Image.open(os.path.join(input_path, fname)).convert('RGB')
    img_array = np.array(image)
    pred_probs = classifier.predict([img_array])  # (1, 1000) probabilities
    pred_index = np.argmax(pred_probs)            # index of highest prob
    true_label = os.path.splitext(fname)[0]

    if pred_index == mapping[true_label]:
        #plt.imshow(image, title='correct prediction')
        print(f"True label: {true_label} (index:{mapping[true_label]}), Predictied Index: {pred_index} --> correct prediction")
    else:
        #plt.imshow(image, title='incorrect prediction')
        print(f"True label: {true_label} (index:{mapping[true_label]}), Predictied Index: {pred_index} --> incorrect prediction")
```
